```python
plt.figure(figsize=(20, 20))
sns.pairplot(df)
plt.show()

# This plot also tells us about the above correlation
```

```python
# Extract y vector from the dataframe

Y = df['Scaled sound pressure level']
X = df.drop('Scaled sound pressure level', axis=1)
# spliting train and test dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=10)
```

```python
l = LinearRegression().fit(X_train, Y_train)

predictions = l.predict(X_test)
```

```python
mse = mean_squared_error(predictions, Y_test)

print("MSE :", mse)
```

```python
# RMSE - It is root of mean of squared errors. It is a metric to evaluate regression tasks. 
# RMSD is always non-negative, and a value of 0 would indicate a perfect fit to the data. 

# Here f is predicted value and o is observed value
display.Image("rmse.png")


```

```python
# MSE - It is an estimator that measures the average of the squares of the errors, that is, 
# the average squared difference between the estimated values and the actual value.
# The MSE is a measure of the quality of an estimator. As it is derived from the square of Euclidean distance, 
# it is always a positive value that decreases as the error approaches zero
display.Image("mse.png")

# Here yi is the observed value and yi^ is the predicted value. n is the total number of samples
```

```python
# MAE is calculated as the sum of absolute errors divided by the sample size.
# It is thus an arithmetic average of the absolute errors

display.Image("mae.png")

# Here yi is the observed value and yi^ is the predicted value. n is the total number of samples
```

```python
def MSE(observed, predicted):
    sum = 0
    if(len(observed)!=len(predicted)):
        print("|observed| != |predicted|")
        return
    for (y_obs,y_pred) in map(lambda x,y:(x,y), observed, predicted):
        sum = sum + (y_obs-y_pred)**2
    sum = sum / len(observed)
    return sum
    
print(MSE([1, 2, 3], [3, 2, 1]))
mse = mean_squared_error([1, 2, 3], [3, 2, 1])
print("MSE :", mse)
```

```python
predictions = l.predict(X_train)
r2 = r2_score(Y_train, predictions)
# R2 value of training data
print(r2)
```

```python
predictions = l.predict(X_test)
r2 = r2_score(Y_test, predictions)
# R2 value of test data
print(r2)
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(df)
plt.show()
```

```python
X = df.loc[:, "Id number":"Fe"]
Y = df["Type of glass"]
# Extract y vector from the dataframe

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=10)
# Spliting train and test dataset
```

```python
logreg = LogisticRegression(random_state=0, multi_class='ovr').fit(X_train, Y_train)
y_pred = logreg.predict(X_test)
```

```python
print("F1 score is : " + str(f1_score(Y_test, y_pred, average="macro")))
print("Precision is : " + str(precision_score(Y_test, y_pred, average="macro")))
print("recall is : " + str(recall_score(Y_test, y_pred, average="macro")))
print("accuracy is : " + str(accuracy_score(Y_test, y_pred)))
```
