```python
import os
import cv2
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
#selecting three random images
files=os.listdir(os.getcwd()+'/MiMM-SBILab')
img1Path=random.choice(files)
files.remove(img1Path)
print(img1Path)
img2Path=random.choice(files)
files.remove(img2Path)
print(img2Path)
img3Path = random.choice(files)
files.remove(img3Path)
print(img3Path)
```

```python
img1 = cv2.imread(os.getcwd()+'/MiMM-SBILab/' + img1Path)
img2 = cv2.imread(os.getcwd()+'/MiMM-SBILab/' + img2Path)
img3 = cv2.imread(os.getcwd()+'/MiMM-SBILab/' + img3Path)
```

```python
#resize
fig = plt.figure(figsize = (10,9))

l1 = int((img1.shape[0])/2)
l2 = int((img1.shape[1])/2)
print('resizing to '+ str(l2) + 'x' + str(l1))
img1_resized = cv2.resize(img1, (l2, l1))
fig.add_subplot(2, 2, 1)
plt.imshow(img3)
plt.axis('off')

print('resizing to '+ str(l2) + 'x' + str(l1))
img1_resized = cv2.resize(img2, (l2, l1))
fig.add_subplot(2, 2, 2)
plt.imshow(img2)
plt.axis('off')

print('resizing to '+ str(l2) + 'x' + str(l1))
img1_resized = cv2.resize(img3, (l2, l1))
fig.add_subplot(2, 2, 3)
plt.imshow(img3)
plt.axis('off')

plt.show()
```

```python
#pad
fig = plt.figure(figsize = (10,9))

image_height, image_width, channels = img1.shape
color = (100,50,100)
result = np.full((image_height + 200, image_width+200, channels), color, dtype=np.uint8)
x_center = 100
y_center = 100
result[y_center:y_center+image_height, x_center:x_center+image_width] = img1
fig.add_subplot(2, 2, 1)
plt.imshow(result)
plt.axis('off')


image_height, image_width, channels = img2.shape
color = (100,50,100)
result = np.full((image_height + 200, image_width+200, channels), color, dtype=np.uint8)
x_center = 100
y_center = 100
result[y_center:y_center+image_height, x_center:x_center+image_width] = img2
fig.add_subplot(2, 2, 2)
plt.imshow(result)
plt.axis('off')

image_height, image_width, channels = img3.shape
color = (100,50,100)
result = np.full((image_height + 200, image_width+200, channels), color, dtype=np.uint8)
x_center = 100
y_center = 100
result[y_center:y_center+image_height, x_center:x_center+image_width] = img3
fig.add_subplot(2, 2, 3)
plt.imshow(result)
plt.axis('off')

plt.show()
```

```python
#crop
fig = plt.figure(figsize = (10,9))

image_height, image_width, channels = img1.shape
result = img1[int(image_height/6):image_height-int(image_height/6),int(image_width/6):image_width-int(image_width/6)]
fig.add_subplot(2, 2, 1)
plt.imshow(result)
plt.axis('off')

image_height, image_width, channels = img2.shape
result = img2[int(image_height/6):image_height-int(image_height/6),int(image_width/6):image_width-int(image_width/6)]
fig.add_subplot(2, 2, 2)
plt.imshow(result)
plt.axis('off')

image_height, image_width, channels = img3.shape
result = img3[int(image_height/6):image_height-int(image_height/6),int(image_width/6):image_width-int(image_width/6)]
fig.add_subplot(2, 2, 3)
plt.imshow(result)
plt.axis('off')

plt.show()
```

```python
#grayscale
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
cv2.imshow("image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
cv2.imshow("image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

gray = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
cv2.imshow("image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#contrast
fig = plt.figure(figsize = (10,9))

alpha = 1.5
beta = 0 
adjusted = cv2.convertScaleAbs(img1, alpha=alpha, beta=beta)
fig.add_subplot(2, 2, 1)
plt.imshow(adjusted)
plt.axis('off')

alpha = 1.5
beta = 0 
adjusted = cv2.convertScaleAbs(img2, alpha=alpha, beta=beta)
fig.add_subplot(2, 2, 2)
plt.imshow(adjusted)
plt.axis('off')

alpha = 1.5
beta = 0 
adjusted = cv2.convertScaleAbs(img3, alpha=alpha, beta=beta)
fig.add_subplot(2, 2, 3)
plt.imshow(adjusted)
plt.axis('off')
plt.show()
```

```python
#saturation
fig = plt.figure(figsize = (10,9))
hsvImage = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
lower_red = np.array([0,10, 0])
upper_red = np.array([255,180,255])
mask = cv2.inRange(hsvImage, lower_red, upper_red)
res = cv2.bitwise_and(img1,img1, mask= mask)
fig.add_subplot(2, 2, 1)
plt.imshow(res)
plt.axis('off')
    
hsvImage = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)
lower_red = np.array([0,10, 0])
upper_red = np.array([255,180,255])
mask = cv2.inRange(hsvImage, lower_red, upper_red)
res = cv2.bitwise_and(img2,img2, mask= mask)
fig.add_subplot(2, 2, 2)
plt.imshow(res)
plt.axis('off')
    
hsvImage = cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)
lower_red = np.array([0,10, 0])
upper_red = np.array([255,180,255])
mask = cv2.inRange(hsvImage, lower_red, upper_red)
res = cv2.bitwise_and(img3,img3, mask= mask)
fig.add_subplot(2, 2, 3)
plt.imshow(res)
plt.axis('off')
plt.show()
```
