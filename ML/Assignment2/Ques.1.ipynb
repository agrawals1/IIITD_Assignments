# Ques-1(a)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("gdp.csv")
df.head()
```

```python
df=df.set_index("States/UTs")
df.head()
```

```python
df.describe()
```

```python
df.info()
#there are some null entries in the dataset
```

```python
df_mean = df.mean(axis = 1).round(2).sort_values()
df_mean
#finding average growth of each state from 1997
```

```python
plt.rcParams['figure.figsize'] = [11.5,8]
df_mean.plot(kind='barh',stacked=True, colormap = 'Set1')
plt.title("Avg.% Growth of States from 1997", fontweight = 'bold')
plt.xlabel("Avg. % Growth", fontweight = 'bold')
plt.ylabel("States/UTs", fontweight = 'bold')
```

```python
df_mean2 = df.iloc[:,-2:].mean(axis = 1).round(2).sort_values()
df_mean2
```

```python
plt.rcParams['figure.figsize'] = [11.5,8]
df_mean2.plot(kind='barh',stacked=True, colormap = 'Set1')
plt.title("Avg.% Growth of States from 2012-2014", fontweight = 'bold')
plt.xlabel("Avg. % Growth", fontweight = 'bold')
plt.ylabel("States/UTs", fontweight = 'bold')
plt.show()
```

```python
df_myhome = df_mean2[['Uttar Pradesh', 'All India']]
```

```python
df_myhome.plot(kind='barh',stacked=True, colormap = 'Dark2')
plt.title("Avg. % Growth of Home State vs National Avg. for Duration 2012-13 and 2013-14", fontweight = 'bold')
plt.ylabel("Average % Growth", fontweight = 'bold')
plt.xlabel("Home State Vs National Average", fontweight = 'bold')
plt.show()
```

```python
df_mean2.tail(5).plot(kind='bar',stacked=True, colormap = 'Dark2')
plt.title("Total GDP of top 5 States for 2012-14", fontweight = 'bold')
plt.ylabel("Total GDP (in cr)",fontweight = 'bold')
plt.xlabel("States",fontweight = 'bold')


```

```python
# GSDP of Bottom 5 States
df_mean2.head(5).plot(kind='bar',stacked=True, colormap = 'Set1')
plt.title("Total GDP of bottom 5 States for 2012-14", fontweight = 'bold')
plt.ylabel("Total GDP (in cr)",fontweight = 'bold')
plt.xlabel("States",fontweight = 'bold')
plt.show()
```

The data is of gdp growth not gdp values. It shows how much gdp of a state as well as all india is growing every year. We have done some analysis of gdp growth over some range of periods from over all dataset. 

The use case of this dataset is that we can find out trends of gdp growth of different states in some year or over some periods and find out that which states and performing less, so that we can put more focus on those states and we can focus less on states which are consistently performing well.

Type of data is suitable for normalization:-
*Normalization is useful when there are no outliers as it cannot cope up with them.It is really affected by outliers.
*It is useful when we don’t know about the distribution or data is not having Gaussian Distribution.
*Normalization scales in a range of [0,1] or [-1,1].

Type of data suitable for standardiztion:-
*Standardization is used when algorithms make assumptions about the data distribution.Standardization is used on data having Gaussian distribution.
*Standardization is useful when there are outliers. It is not affected by outliers.
*Standardization is not bounded by range.


From above two, standardization would be a better choice on given data set. It's because the gdp growth shows somewhat a gaussian distribution, having values starting with negative, attaining a max. value and then decreasing.
